```python
# Import tkinter and time libraries:

import tkinter as tk
from tkinter import *
import time

# Taking input from user to validate which background color they want:

background_color = input("Enter your favourite background color:")

# Taking input from user to validate which text color they want:

color = input("Enter your favurite text color:")

# Initialising the GUI: 

root = tk.Tk()
root.geometry("700x600")
root.configure(background="lightgreen")
root.title("Namaz Clock")
 


# Defining Function to make the clock show the time in realtime: 

def tick(time1 = ""):
    time2 = time.strftime('%I:%M:%S')
    clock_frame.config(text = time2)
    clock_frame.after(20, tick)

# Function to display Fajr namaz time:
    
def Fajar_namaz():
    Fajar.config(text = "Fajar: 05:40 a.m")
    
# Function to display Zuhar namaz time:    
    
def Zuhar_namaz():
    Zuhar.config(text = "Zuhar: 01:30 p.m")
    
# Function to display Asar namaz time:    
    
def Asar_namaz():
    Asar.config(text = "Asar: 04:20 p.m")
    
# Function to display Maghrib namaz time:    
    
def Maghrib_namaz():
    Maghrib.config(text = "Maghrib: 06:55 p.m")
    
# Function to display Isha namaz time:    
    
def Isha_namaz():
    Isha.config(text = "Isha: 08:15 p.m")
    
# Function to display Jumma namaz time:    
    
def Jumma_namaz():
    Jumma.config(text = "Jumma: 02:00 p.m ")
    
# Designing the clock showing the time in realtime:

clock_frame= tk.Label(root, font=('arial',100, 'bold'), bg =background_color, fg = color)
clock_frame.pack(fill = "both", expand = 1)

# Designing the clock showing the time in realtime:

# Designing the clock showing the Fajr Namaz time:

Fajar= tk.Label(root, font=('arial',20, 'bold'), bg =background_color, anchor = 'n', fg = color)
Fajar.pack(fill = X)


# Designing the clock showing the Zuhar Namaz time:

Zuhar= tk.Label(root, font=('arial',20, 'bold'), bg =background_color, anchor = 'n', fg = color)
Zuhar.pack(fill = X)


# Designing the clock showing the Asar Namaz time:

Asar= tk.Label(root, font=('arial',20, 'bold'), bg =background_color, anchor = 'n', fg = color)
Asar.pack(fill = X)


# Designing the clock showing the Maghrib Namaz time:

Maghrib= tk.Label(root, font=('arial',20, 'bold'), bg =background_color, anchor = 'n', fg = color)
Maghrib.pack(fill = X)


# Designing the clock showing the Isha Namaz time:

Isha= tk.Label(root, font=('arial',20, 'bold'), bg =background_color, anchor = 'n', fg = color)
Isha.pack(fill = X)


# Designing the clock showing the Jumma Namaz time:

Jumma= tk.Label(root, font=('arial',20, 'bold'), bg =background_color, anchor = 'n', fg = color)
Jumma.pack(fill = X)

# Calling the function to display current time in real time:

tick()

# Calling the function to display current Fajar Namaz time:

Fajar_namaz()

# Calling the function to display current Zuhar Namaz time:

Zuhar_namaz()


# Calling the function to display current Asar Namaz time:

Asar_namaz()


# Calling the function to display current Maghrib Namaz time:

Maghrib_namaz()


# Calling the function to display current Isha Namaz time:

Isha_namaz()


# Calling the function to display current Jumma Namaz time:

Jumma_namaz()

# Starting the GUI:

root.mainloop()
```
